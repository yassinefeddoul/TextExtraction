```python
import pytesseract
import cv2
import PIL
```

```python
def grayscale(image):
    return cv2.cvtColor(cv2.UMat(image), cv2.COLOR_BGR2GRAY)
 
# noise removal
def remove_noise(image):
    return cv2.medianBlur(cv2.UMat(image),5)
  
# thresholding
def thresholding(image):
    return cv2.threshold(cv2.UMat(image), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
```

```python
image_path = r'C:\Users\dell\Downloads\bayer.jpg'
im = cv2.imread(image_path)
print(im)
```

```python
PIL.Image.open(image_path)
```

```python
%%time
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

imgg=remove_noise(im)
img= cv2.UMat.get(imgg)
text =pytesseract.image_to_string(img, lang="eng")
```
